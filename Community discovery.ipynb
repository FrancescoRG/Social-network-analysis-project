```python
#Importo le librerie necessarie allo svolgimento del task
import networkx as nx
import pquality
from networkx.algorithms import community
import matplotlib.pyplot as plt
from networkx.algorithms.community import kclique as kq
from networkx.algorithms.community import label_propagation as label
from networkx.algorithms.community.centrality import girvan_newman as gn
from pquality.PartitionQuality import pquality_summary
import community as louvain


def takeSecond(elem):
    return elem[1]

#Genero un grafico vuoto che andrò successivamente a riempire
Graphtype=nx.Graph()   

#Leggo dal file txt la lista dei link che compongono il grafico
G = nx.read_edgelist('Speranza.txt', create_using=Graphtype, nodetype=int)
print(nx.info(G))

#richiamo dalla libreria l'algoritmo k-clique con k=4
kclique4 = list(community.k_clique_communities(G, 4))
print('Comunità individuate: ' + str(len(kclique4)))
print(kclique4)



#Stampo un sommario utile a riassumere le misure registrate
results = pquality.pquality_summary(G, kclique4) 
print(results['Indexes'])
print(results['Modularity'])


#Valuto la presenza di nodi in overlap
#dizionario che ha come etichetta i nodi e come valore le loro occorrenze nelle comunità
conteggio={} 

for kcom in kclique4:
    for k in kcom:
        if k in conteggio.keys():
            conteggio[k] += 1
        else:
            conteggio[k] = 1

conta=0 #conta il numero di nodi che compaiono in più di una comunità

for k, v in sorted(conteggio.items(), key=takeSecond, reverse=True):
    print(k,v)
    if (v>1):
        conta+=1
        
print("I nodi che fanno overlapping sono "+str(conta))
```

```python
#Louvain
import community as louvain
from collections import defaultdict
coms = louvain.best_partition(G)
# E' necessario rimodellare i risultati per renderli nello stesso formato degli altri algoritmi per la CD 
coms_to_node = defaultdict(list)
for n, c in coms.items():
    coms_to_node[c].append(n)

coms_louvain = [tuple(c) for c in coms_to_node.values()]
print('Comunità individuate: ' + str(len(coms_louvain)))
print(coms_louvain)
results = pquality.pquality_summary(G, coms_louvain) 
print(results['Indexes'])
print(results['Modularity'])
#Valuto la presenza di nodi in overlap
#dizionario che ha come etichetta i nodi e come valore le loro occorrenze nelle comunità
conteggio={} 

#Questo algoritmo non fa overlapping perchè top-down
for kcom in coms_louvain:
    for k in kcom:
        if k in conteggio.keys():
            conteggio[k] += 1
        else:
            conteggio[k] = 1

conta=0 #conta il numero di nodi che compaiono in più di una comunità

for k, v in sorted(conteggio.items(), key=takeSecond, reverse=True):
    print(k,v)
    if (v>1):
        conta+=1
print("I nodi che fanno overlapping sono "+str(conta))
```
